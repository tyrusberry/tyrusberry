```python
import numpy as np
import random as random
import math
from math import comb

def c_norm(a,alpha,entry):
    A = np.array([[comb(alpha,i)*a[j]**(alpha-i)/((a[j]**2+1)**(alpha/2)) for j in range(alpha+1)] for i in range(alpha+1)])
    b = np.array([0 for i in range(alpha+1)])
    b[entry]=1

    x = np.linalg.solve(A, b)

    return sum(abs(i) for i in x)
```

```python
import scipy.optimize as op

for alpha in range(3, 15):
    print('For the ' + str(alpha) + '-tensor:')
    entry = 2

    best_aopt = []
    best_copt = 999

    for num in range(10):
        a0 = [random.random() for i in range(alpha + 1)]
        res = op.minimize(c_norm, a0, method='Nelder-Mead', tol=1e-6, options={'maxiter': 5000}, args=(alpha, entry))
        aopt = res.x
        copt = res.fun
        if copt < best_copt:
            best_copt = copt
            best_aopt = aopt
    print('Optimal a vector:', np.round(best_aopt, 3))
    print('Optimal Coefficient:', np.round(best_copt, 3))
    print('\n')
```
